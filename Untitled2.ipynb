```python
x = df["YEAR"]
y = df["JAN"]
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8,random_state=4)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train.values.reshape(-1,1),y_train)
```

```python
y_pred = lr.predict(x_test.values.reshape(-1,1))
```

```python
y_pred
```

```python
rsquare = lr.score(x_train.values.reshape(-1,1), y_train)
```

```python
rsquare
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
mean_absolute_error(y_test, y_pred)
```

```python
mean_squared_error(y_test, y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```

```python
plt.scatter(x_train,y_train,color='Black')
plt.plot(x_train,lr.predict(x_train.values.reshape(-1,1)),color='blue',linewidth=3)
plt.title("Temperature vs. Year")
plt.xlabel("YEAR")
plt.ylabel("TEMPERATURE")
plt.show()
```

```python
plt.scatter(x_test,y_test,color='Red')
plt.plot(x_test,lr.predict(x_test.values.reshape(-1,1)),color='Green',linewidth=3)
plt.title("Temperature vs. Year")
plt.xlabel("YEAR")
plt.ylabel("TEMPERATURE")
plt.show()
```
